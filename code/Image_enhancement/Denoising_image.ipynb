### Gaussian denoising, experiment with sigma

```python
g_denoise_gaussian_3x3_s0_5 = denoise(img1_gaussian, 'gaussian', ksize=(3,3), sigma=0.5)
plt.imshow(g_denoise_gaussian_3x3_s0_5)
```

```python
g_denoise_gaussian_3x3_s1 = denoise(img1_gaussian, 'gaussian', ksize=(3,3), sigma=1)
plt.imshow(g_denoise_gaussian_3x3_s1)
```

```python
g_denoise_gaussian_3x3_s1_5 = denoise(img1_gaussian, 'gaussian', ksize=(3,3), sigma=1.5)
plt.imshow(g_denoise_gaussian_3x3_s1_5)
```

```python
g_denoise_gaussian_3x3_s2 = denoise(img1_gaussian, 'gaussian', ksize=(3,3), sigma=2)
plt.imshow(g_denoise_gaussian_3x3_s2)
```

```python
g_denoise_gaussian_3x3_s3 = denoise(img1_gaussian, 'gaussian', ksize=(3,3), sigma=3)
plt.imshow(g_denoise_gaussian_3x3_s3)
```

```python
print("PSNR g gaussian 3x3 sigma 0.5: ", myPSNR(img1, g_denoise_gaussian_3x3_s0_5))
print("PSNR g gaussian 3x3 sigma 1: ", myPSNR(img1, g_denoise_gaussian_3x3_s1))
print("PSNR g gaussian 3x3 sigma 1.5: ", myPSNR(img1, g_denoise_gaussian_3x3_s1_5))
print("PSNR g gaussian 3x3 sigma 2: ", myPSNR(img1, g_denoise_gaussian_3x3_s2))
print("PSNR g gaussian 3x3 sigma 3: ", myPSNR(img1, g_denoise_gaussian_3x3_s3))
```
